## PART 1

```python
# Importing the packages and the raw data

import pandas as pd
import numpy as np

link = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

DF = pd.read_html(link)[0]
DF
```

```python
# Ignore cells with a borough that is Not assigned.

DF = DF[DF['Borough'] != 'Not assigned']
DF
```

```python
# Combining neighbourhoods into one row with the neighborhoods separated with a comma

DF['Neighbourhood'] = DF.groupby("Postcode")["Neighbourhood"].transform(lambda neigh: ', '.join(neigh))
DF.drop_duplicates(subset ="Postcode", inplace = True)
DF
```

```python
#If a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough.

DF['Neighbourhood'].replace("Not assigned", DF["Borough"],inplace=True)
DF
```

### I decided to work with only boroughs that contain the word Toronto

```python
toronto_df = DF[DF['Borough'].str.contains('Toronto', na = False)]
toronto_df.reset_index(drop = True, inplace = True)
toronto_df.head()
```

```python
# Creating a map of Toronto.

lat = 43.6532
long = -79.3832

map_toronto = folium.Map(location=[lat, long], zoom_start=11)

for lat, long, label in zip(toronto_df['Latitude'], toronto_df['Longitude'], toronto_df['Neighbourhood']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, long],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
